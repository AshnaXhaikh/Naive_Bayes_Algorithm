# **Exploring and Visualizing a Simple Dataset**

```python
# importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Load the **iris dataset** through sk-learn library.

```python
# Load Iris dataset using sklearn and convert to DataFrame
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
X = iris.data # features
y = iris.target # labels
```

```python
# create a Dataframe
df = pd.DataFrame(X, columns=iris.feature_names)
df['species'] = y
df.head()
```

## 2. Display Dataset Structure

```python
print("Feature shape:", X.shape)
print("Target shape:", y.shape)
```

```python
print(f"Shape of iris: {df.shape}")
```

```python
features = df.columns.tolist()
print("column names: \n", ", ".join(features))
```

```python
labels = iris.target_names
print("Labels: \n", ", ".join(labels.tolist()))
```

### Convert numeric target labels to species names for easier analysis and plotting.

```python
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
df.sample()
```

```python
df.head()
```

# **Step 3: Basic Visualizations**

# Scatter Plot â€” Relationship Between Variables

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='sepal length (cm)', y='sepal width (cm)', hue='species')
plt.title('=Sepal Length vs. Sepal Width')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='petal length (cm)', y='petal width (cm)', hue='species')
plt.title('Petal Length vs. Petal Width')
plt.show()
```

```python

# Pairwise scatter plots colored by species
sns.pairplot(df, hue='species', diag_kind='hist')
plt.suptitle("Pairwise Scatter Plot of Iris Features", y=1.02)
plt.show()
```

# Histogram â€” Data Distribution

```python
# Histograms for each feature
df.iloc[:, :-1].hist(bins=20, figsize=(10, 8))
plt.suptitle("Histograms of Iris Features")
plt.show()
```

# histogram by species

```python
for col in features:
  plt.figure(figsize=(6, 4))
  sns.histplot(data=df, x=col, hue='species', multiple='stack', bins=20)
  plt.title(f"Distribution of {col} by Species")
  plt.show()
```

# **Box Plot â€” Outliers and Spread**
  - to Detect Outliers and Spread of Values.
  - Distribution and spread of each feature within each species.
  - Species differences in median, range, and outliers.
  
  - Each label tells you which feature that particular box plot is summarizing.

  - Each box plot summarizes the spread and outliers of that feature across all flowers.


# Simple matplotlib histograms

```python
# Combined box plots for all features
plt.figure(figsize=(12, 8))
sns.boxplot(data=df.iloc[:, :-1])
plt.title('Box Plots of Iris Features')
plt.xlabel('Features')
plt.ylabel('Value (cm)')
plt.show()
```

# Seaborn histogram by species

```python
for col in iris.feature_names:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x='species', y=col, data=df)
    plt.title(f'Box Plot of {col} by Species')
    plt.xlabel('Species')
    plt.ylabel(f'{col} (cm)')
    plt.show()
```
